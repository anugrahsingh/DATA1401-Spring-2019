### Heart Data

[Source](https://www.kaggle.com/ronitf/heart-disease-uci/version/1#)

* *age*- age in years
* *sex*- (1 = male; 0 = female)
* *cp*- chest pain type
* *trestbps*- resting blood pressure (in mm Hg on admission to the hospital)
* *chol*- serum cholestoral in mg/dl
* *fbs*- (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* *restecg*- resting electrocardiographic results
* *thalach*- maximum heart rate achieved
* *exang*- exercise induced angina (1 = yes; 0 = no)
* *oldpeak*- ST depression induced by exercise relative to rest
* *slope*- the slope of the peak exercise ST segment
* *ca*- number of major vessels (0-3) colored by flourosopy
* *thal*- 3 = normal; 6 = fixed defect; 7 = reversable defect
* *target*- 1 or 0

```python
np_heart=np.array(df_heart)
np_heart.shape
```

```python
h_c_tmp=df_heart.columns.tolist()
h_columns=dict(zip(h_c_tmp,range(len(h_c_tmp))))
h_columns
```

```python
index=np.where(np_heart[:,h_columns["cp"]]==0.)
```

```python
np_heart[index,h_columns["chol"]].flatten().shape
```

```python
_=plt.hist(np_heart[index,h_columns["chol"]].flatten(),bins=50)
```

```python
index=np.where(np_heart[:,h_columns["cp"]]==0)
_=plt.hist(np_heart[index,h_columns["chol"]].flatten(),color="r",bins=50,normed=1.)
plt.hold("on")
index=np.where(np_heart[:,h_columns["cp"]]!=0)
_=plt.hist(np_heart[index,h_columns["chol"]].flatten(),color="b",bins=50,normed=1.)
```
